```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.cluster import KMeans, SpectralClustering
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
X, y = make_classification(n_samples=200, n_features=10,n_classes=4,n_informative=4)
```

```python
X.shape
```

```python
y.shape
```

```python
np.unique(y, return_counts=True)
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
pred = kmeans.fit(X)
```

```python
predicted = pred.labels_
```

```python
accuracy_score(y, predicted)
```

```python
SkMeans = SpectralClustering(n_clusters=4)
```

```python
Spred = SkMeans.fit(X)
```

```python
SPrediction = Spred.labels_
```

```python
accuracy_score(y, SPrediction)
```

```python
pca = PCA(n_components=2)
```

```python
X_trans = pca.fit_transform(X)
```

```python
X_trans.shape
```

```python
plt.scatter(X_trans[:,0], X_trans[:,1], c=y)
plt.title("Original Dataset")
plt.show()
```

```python
plt.scatter(X_trans[:,0], X_trans[:,1], c=predicted)
plt.title("KMeans")
plt.show()
```

```python
plt.scatter(X_trans[:,0], X_trans[:,1], c=SPrediction)
plt.title("Spherical KMeans")
plt.show()
```
